## while 循环
- 当一个条件保持真的时候while循环重复执行语句
- while 循环一定要有结束条件，否则很容易进入死循环
- while 循环的语法是：

    while loop-contunuation-conndition:

        Statement

## 示例：
sum = 0

i = 1

while i <10:

    sum = sum + i
    i = i + 1

## 错误示例：

    sum = sum + i

## 嵌套循环
- 一个循环可以嵌套另一个循环
- 每次循环外层时，内层循环都会被刷新重新完成循环
- 也就是说，大循环执行一次，小循环会全部执行一次
- 注意：
> - 多层循环非常耗时
  - 最多使用3层循环

## EP：
- 使用多层循环完成9X9乘法表
- 显示50以内所有的素数

```python
for i in range(1,10):
    for j in range(i,10):
        s = i * j
        print(i,'*',j,'=',s,end=' ')
        if j==9 :
            print(end='\n')
```

```python
for i in range(2,50):
    if i == 2:
        print(i,end=' ')
    for j in range(2,i):
        if i % j == 0:
            break
        else:
            print(i,end=' ')
            break
```
